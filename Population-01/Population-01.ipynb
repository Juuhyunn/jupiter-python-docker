```python
import csv
import matplotlib.pyplot as plt
import numpy
```

```python
data = csv.reader(open('./data/202106_202106_연령별인구현황_월간.csv', encoding='utf-8'))
next(data)
data = list(data)
```

```python
for i in data:
    if '방이1동' in i[0]:
        arr = numpy.array(i[3:], dtype=int) / int(i[2].replace(',', ''))
```

```python
plt.title('Population-01')
plt.style.use('ggplot')
plt.plot(arr)
```
